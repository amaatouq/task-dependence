```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import lognorm
sns.set(style="white", context="talk")
from matplotlib import rc

rc('text', usetex=True)
```

```python
def ωandn(ω,n):
    w = (1-ω)*np.ones(n)/n
    w[0]+=ω
    return w
```

```python
θ = 2
N = 10000
n = 50
```

```python
ω = 1/3
w_star = ωandn(ω,n)
w_cycle = ωandn(0,n)
```

```python
μ = -1
σ = 1

Lognormal = lognorm(s=σ,scale=np.exp(μ+np.log(θ)))

a_star = []
a_cycle = []

for i in range(N):
    signals = Lognormal.rvs(size=n)
    a_star.append(w_star@signals)
    a_cycle.append(w_cycle@signals)
```

```python
plt.figure(figsize=(12,8))
sns.distplot(a_star,label=r"$\mathbf{a}^{n}(\omega)$")
sns.distplot(a_cycle,label=r"$\mathbf{a}^{n}(0)$")
plt.xlim(0,2*θ)
plt.legend()
```

```python
plt.figure(figsize=(12,8))

x = np.linspace(0,2*θ,num=400)
plt.plot(x,Lognormal.pdf(x),alpha=0.5)
#plt.axvline(Lognormal.mean(), color='r',label=r"population mean")
#plt.axvline(Lognormal.median(), color='b',label=r"population median")
plt.scatter(Lognormal.median(),0., color='b', marker='^',s=1000)
plt.scatter(Lognormal.mean(),0., color='r', marker='^',s=1000)
plt.axvline(θ,color='k',linewidth=3,linestyle='--',label=r"$\theta$")
plt.axvline(np.median(a_star),color='b',linestyle='--',label=r"median of $\mathbf{a}^{n}(\omega)$")
plt.axvline(np.median(a_cycle),color='r',linestyle='--',label=r"median of $\mathbf{a}^{n}(0)$")
plt.ylim(0,1)
plt.xlim(0,2*θ)
plt.yticks([])
plt.title(r"$\mu={}$, $\sigma={}$, $\theta={}$".format(μ,σ,θ),fontsize=30)
plt.legend()
plt.savefig('mu_small.pdf', bbox_inches='tight')
```
